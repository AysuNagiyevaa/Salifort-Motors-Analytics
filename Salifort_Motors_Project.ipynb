```python
# Create a plot as needed
### YOUR CODE HERE ###
# Donut chart: overall turnover split by salary level
salary_turnover = pd.crosstab(df['salary'], df['left'])
salary_turnover = salary_turnover.rename(columns={0: 'Stayed', 1: 'Left'})

fig, axes = plt.subplots(1, len(salary_turnover.index), figsize=(12, 4))
for ax, salary in zip(axes, salary_turnover.index):
    values = salary_turnover.loc[salary]
    ax.pie(values, labels=values.index, autopct='%1.1f%%', startangle=90, wedgeprops={'width': 0.45})
    ax.set_title(f'{salary.capitalize()} Salary')

plt.suptitle('Turnover Share Within Each Salary Level', fontsize=14)
plt.tight_layout()
plt.show()
```

```python
# Create a plot as needed
### YOUR CODE HERE ###
# Line chart: turnover rate by tenure
tenure_rate = df.groupby('tenure')['left'].mean().reset_index()
tenure_rate['turnover_rate_percent'] = tenure_rate['left'] * 100

plt.figure(figsize=(8, 5))
sns.lineplot(data=tenure_rate, x='tenure', y='turnover_rate_percent', marker='o')
plt.title('Turnover Rate Trend by Tenure')
plt.xlabel('Tenure (Years)')
plt.ylabel('Turnover Rate (%)')
plt.ylim(0, tenure_rate['turnover_rate_percent'].max() + 10)
plt.grid(axis='y', alpha=0.3)
plt.show()
```

```python
# Create a plot as needed
### YOUR CODE HERE ###
# Horizontal lollipop chart: turnover rate by department
dept_rate = (df.groupby('department')['left']
             .mean()
             .mul(100)
             .sort_values()
             .reset_index())

plt.figure(figsize=(9, 6))
plt.hlines(y=dept_rate['department'], xmin=0, xmax=dept_rate['left'])
plt.plot(dept_rate['left'], dept_rate['department'], 'o')
plt.title('Department Turnover Rate Ranking')
plt.xlabel('Turnover Rate (%)')
plt.ylabel('Department')
plt.grid(axis='x', alpha=0.3)
plt.show()
```

```python
# Create a plot as needed
### YOUR CODE HERE ###
# Heatmap: average turnover rate by satisfaction and evaluation groups
df_heatmap = df.copy()
df_heatmap['satisfaction_group'] = pd.cut(
    df_heatmap['satisfaction_level'],
    bins=[0, 0.35, 0.70, 1.00],
    labels=['Low', 'Medium', 'High'],
    include_lowest=True
)
df_heatmap['evaluation_group'] = pd.cut(
    df_heatmap['last_evaluation'],
    bins=[0, 0.50, 0.75, 1.00],
    labels=['Low', 'Medium', 'High'],
    include_lowest=True
)

heatmap_data = df_heatmap.pivot_table(
    index='satisfaction_group',
    columns='evaluation_group',
    values='left',
    aggfunc='mean',
    observed=False
) * 100

plt.figure(figsize=(7, 5))
sns.heatmap(heatmap_data, annot=True, fmt='.1f', cmap='YlOrRd', cbar_kws={'label': 'Turnover Rate (%)'})
plt.title('Turnover Risk by Satisfaction and Evaluation Groups')
plt.xlabel('Final Evaluation Group')
plt.ylabel('Satisfaction Group')
plt.show()
```

## Recall evaluation metrics

- **AUC** is the area under the ROC curve; it's also considered the probability that the model ranks a random positive example more highly than a random negative example.
- **Precision** measures the proportion of data points predicted as True that are actually True, in other words, the proportion of positive predictions that are true positives.
- **Recall** measures the proportion of data points that are predicted as True, out of all the data points that are actually True. In other words, it measures the proportion of positives that are correctly classified.
- **Accuracy** measures the proportion of data points that are correctly classified.
- **F1-score** is an aggregation of precision and recall.
